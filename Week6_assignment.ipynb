# Feature Engineering

```python
cat_colm=df.select_dtypes(include=['object']).columns
num_colm=df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables are:",cat_colm)
print("Numerical Variables are:",num_colm)
```

```python
for col in num_colm:
    print(col)
    plt.figure(figsize=(16,5))
    plt.subplot(1,2,1)
    sns.histplot(df[col],kde=True)
    plt.ylabel("Count")
    plt.subplot(1,2,2)
    sns.boxplot(df[col])
    plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
for col in df.columns:
    if col != "Total Revenue":
        plt.figure(figsize=(8,4))
        sns.countplot(x=col, hue="Total Revenue", data=df)
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.title(f"{col} vs Total Revenue")
        plt.show()
```
